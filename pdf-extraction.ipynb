### 1. Using PyPDF

### Analysing outputs of extraction for different PDFs

```python
from PyPDF2 import PdfReader
def load_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text
print(load_pdf('data/insurance.pdf')) # guardian vision insurance
```

## 3. Docling

### Using default mode: maps structure back to pdf cells
- Identifies where each cell starts and ends based on the visual layout in the PDF
- This mode is ideal when:
    - Tables are clearly formatted
    - Columns and rows are well-alligned
    - Borders and cell boundaries are visible

If the PDF has:
- Merged cells or columns that are incorrectly aligned.
- Tables without clear borders.
- Irregular spacing, making it hard to determine where a cell ends and another begins.

Then the default mode might extract text inaccurately, such as:
- Multiple columns merging into one.
- Missing rows or incomplete cells.

### Adjusting pipeline features
- To control if table structure recognition should map the recognized structure back to PDF cells (default) or use text cells from the structure prediction itself
- This can improve output quality if you find that multiple columns in extracted tables are erroneously merged into one.


```python
from docling.datamodel.base_models import InputFormat
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions

pipeline_options = PdfPipelineOptions(do_table_structure=True)
pipeline_options.table_structure_options.do_cell_matching = False  # uses text cells predicted from table structure model

doc_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)

result = converter.convert(source)
print(result.document.export_to_markdown())
```

### Tableformer - FAST(default) and ACCURATE

```python
from docling.datamodel.base_models import InputFormat
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions, TableFormerMode

pipeline_options = PdfPipelineOptions(do_table_structure=True)
pipeline_options.table_structure_options.mode = TableFormerMode.ACCURATE  # use more accurate TableFormer model

doc_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)

result = doc_converter.convert(source)
print(result.document.export_to_markdown())
```

### Chunking docling document

A docling document can be chunked using a chunker such as HybridChunker

```python
from docling.document_converter import DocumentConverter
from docling.chunking import HybridChunker

conv_res = DocumentConverter().convert("data/insurance.pdf")
doc = conv_res.document

chunker = HybridChunker(tokenizer="BAAI/bge-small-en-v1.5",max_chunk_size=512)  # set tokenizer as needed
chunk_iter = chunker.chunk(doc)
```

### Extracting PDFs using Megaparse

```python
import asyncio
import pytesseract
# setting path to resolve tesseract not in path error
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

model = ChatOpenAI(model='gpt-4', api_key=os.getenv("OPENAI_API_KEY"))
parser = UnstructuredParser()
megaparse = MegaParse(parser)

# async wrapper
async def parse_file(parser, file_path):
    return await parser.convert(file_path)

# parsing
async def main():
    response = await parse_file(parser, "data/insurance.pdf")
    print(response)

asyncio.run(main())
```
